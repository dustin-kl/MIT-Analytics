```python
from http import client
from urllib import response
import urllib.request
import requests
import pandas as pd
import json
import pprint
import seaborn as sns
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
baseURL = "https://api.rarible.org/v0.1/items/all?showDeleted=false&size=1000"
currentURL = f"{baseURL}"
continuationToken = ""

targetNFTs = 269000
totalNFTs = 0

allNFTs = []

while (totalNFTs < targetNFTs):
    apiJsonRet = requests.get(currentURL).json()

    if ("code" in apiJsonRet.keys() and apiJsonRet["code"] == "UNKNOWN"):
        print("Unknown Code")
        continue

    continuationToken = apiJsonRet["continuation"]
    allNFTs.extend(apiJsonRet["items"])

    totalNFTs += len(apiJsonRet["items"])
    print("total NFTs so far: ", totalNFTs)
    
    currentURL = f"{baseURL}&continuation={continuationToken}"
```

```python
listOfKeywords = ["hoop", "dunk", "basket", "rebound", "nba", "slam dunk", "jump shot", "basketball", "courtside", "celtics", "golden state warriors", "miami heat", "lebron", "stephen curry", "kobe"]

counter = 0
counterNoMeta = 0
counterMatches = 0

NFTsWithMatches = []

for NFTJson in allNFTs:
    counter += 1

    if ("meta" not in NFTJson.keys()):
        counterNoMeta += 1
        continue

    metaData = NFTJson["meta"]
    metaDataString = str(metaData).lower()

    booleanList = [(x in metaDataString) for x in listOfKeywords]
    numberMatches = len([1 for x in booleanList if x == True])

    if (numberMatches > 0):
        counterMatches += 1
        NFTsWithMatches.append(NFTJson)

print("Counter: ", counter, "\nNo Metadata Matches: ", counterNoMeta, "\nNumber of Keyword Matches: ", counterMatches)
```

```python
list_of_owners = []
#baseURLAcc = "https://api.rarible.org/v0.1/ownerships/byItem/itemId="
baseURLAcc = "https://eth-mainnet.g.alchemy.com/nft/v2/demo/getOwnersForToken?" 
currentURLAcc = ""
counterNoAnswer = 0

addressDict = {}

for NFTMatch in NFTsWithMatches:

    if ("blockchain" not in NFTMatch.keys() or "contract" not in NFTMatch.keys() or "tokenId" not in NFTMatch.keys()):
        counterNoAnswer += 1
        continue

    blockchain = NFTMatch["blockchain"]
    chainAndContract = NFTMatch["contract"]
    contractID = chainAndContract.partition(":")[2]
    tokendId = NFTMatch["tokenId"]

    #print(currentURLAcc)

    #currentURLAcc = f"{baseURLAcc}{blockchain}%3A{contractID}%3A{tokendId}"
    currentURLAcc = f"{baseURLAcc}contractAddress={contractID}&tokenId={tokendId}"
    try:
        apiJsonRetAcc = requests.get(currentURLAcc).json()
    except:
        print("Couldnt Load JSON")
        continue

    if ("owners" not in apiJsonRetAcc.keys()):
        counterNoAnswer += 1
        continue

    holderAddresses = apiJsonRetAcc["owners"]
    
    for add in holderAddresses:
        if (add in addressDict.keys()):
            addressDict[add] += 1
        else:
            addressDict[add] = 1

    list_of_owners.extend(holderAddresses)
    list_of_owners = list(set(list_of_owners))
```

```python
etherScanKey = "SMSVJJNKN8SWT2R6G5MSDB3FRYSDYNY7Y3"
etherScanBaseURL = "https://api.etherscan.io/api?module=account&action=balancemulti"

chunksOfAddresses = [list_of_owners[x:x+20] for x in range(0, len(list_of_owners), 20)]

for chunk in chunksOfAddresses:

   separatedAddressesofChunk = ",".join(chunk)
   etherScanCurrentURL = f"{etherScanBaseURL}&address={separatedAddressesofChunk}&tag=latest&apikey={etherScanKey}"

   apiJsonRetBal = requests.get(etherScanCurrentURL).json()
   if ("result" not in apiJsonRetBal.keys()):
      continue

   balances = apiJsonRetBal["result"]

   for i in balances:

      if ("balance" not in i.keys() or "account" not in i.keys()):
         continue

      balance = i["balance"]
      account = i["account"]

      addressDict[account] = (addressDict[account], float(balance) * (10**(-18)))
   
```
